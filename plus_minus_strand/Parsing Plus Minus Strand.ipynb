```python
import os
import numpy as np
import matplotlib.pyplot as plt
import pickle
from collections import defaultdict
```

```python
files = ['plus_minus_simulation/'+file for file in os.listdir('plus_minus_simulation/') if 'results' in file]
files_merck = ['plus_minus_simulation_low_mrate/'+file for file in os.listdir('plus_minus_simulation_low_mrate/') if 'results' in file]
```

```python
all_results = defaultdict(list)

for file in files:
    
    index_progression = int(file.split('minus_')[-1].split('_')[0])
    
    with open(file, 'rb') as f:
        res = pickle.load(f)
    
    all_results[index_progression] += res
```

```python
all_results_merck = defaultdict(list)

for file in files_merck:
    
    index_progression = int(file.split('minus_')[-1].split('_')[0])
    
    with open(file, 'rb') as f:
        res = pickle.load(f)
    
    all_results_merck[index_progression] += res
```

```python
fig, ax = plt.subplots(2,2,figsize = (15,10))

ax[0,0].hist(np.array(all_results[0])[:,1], range = (0,150), bins = 150, histtype = 'step')
ax[0,1].hist(np.array(all_results[4])[:,1], range = (0,150), bins = 150, histtype = 'step')
ax[1,0].hist(np.array(all_results[2])[:,1], range = (0,150), bins = 150, histtype = 'step')
ax[1,1].hist(np.array(all_results[3])[:,1], range = (0,150), bins = 150, histtype = 'step')

ax[0,0].hist(np.array(all_results_merck[0])[:,1], range = (0,150), bins = 150, histtype = 'step')
ax[0,1].hist(np.array(all_results_merck[4])[:,1], range = (0,150), bins = 150, histtype = 'step')
ax[1,0].hist(np.array(all_results_merck[2])[:,1], range = (0,150), bins = 150, histtype = 'step')
ax[1,1].hist(np.array(all_results_merck[3])[:,1], range = (0,150), bins = 150, histtype = 'step')

ax[0,0].set_title('1 -> 1 -> 100', fontsize = 16)
ax[0,1].set_title('1 -> 10 -> 100', fontsize = 16)
ax[1,0].set_title('1 -> 25 -> 100', fontsize = 16)
ax[1,1].set_title('1 -> 100 -> 100', fontsize = 16)

ax[0,0].legend(['$\mu$ = $10^{-6}$', '$\mu$ = $10^{-5}$'], fontsize = 15)

for i in range(2):
    ax[1,i].set_xlabel('Number of mutants in patient', fontsize = 14)
    for j in range(2):
        ax[i,j].tick_params(labelsize = 14)
        ax[i,j].yaxis.offsetText.set_fontsize(13)

plt.tight_layout()
plt.savefig('results_plus_minus_strands.jpg', dpi = 600, bbox_inches = 'tight')
```
