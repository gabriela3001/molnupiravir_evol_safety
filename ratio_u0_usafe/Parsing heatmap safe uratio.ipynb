```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import pickle
import matplotlib
from numpy import nan
```

```python
files = [file for file in os.listdir('treatment_peak') if 'results' in file]
```

```python
a1values = list(np.linspace(7.7,10,800))
mvalues = list(np.linspace(1500,25000,800))[::-1]
```

```python
matrix_heatmap = np.zeros((len(mvalues), len(a1values)))

for find in range(len(files)):
    if find % 100 == 0:
        print(find)
    with open('treatment_peak/'+files[find], 'rb') as f:
        file_result = pickle.load(f)
        
    for r in file_result:
        mind = mvalues.index(r['m'])
        a1ind = a1values.index(r['a1'])
        matrix_heatmap[mind, a1ind] =  r['u_ratio']
```

```python
def calculate_ticks(ticklabels, truerange, tickrange):
    tick_pos = []
    for l in ticklabels:
        NewValue = (((l - truerange[0]) * (tickrange[1] - tickrange[0])) / (truerange[1] - truerange[0])) + tickrange[0]
        tick_pos.append(NewValue)
    return(np.array(tick_pos))
```

```python
param_ticks = {'m':[1500,5000,10000,15000,20000,25000],
               'a1':[7.7,8.0,8.3,8.6,8.9,9.2,9.5,9.8,10]}
```

```python
fig, ax = plt.subplots(figsize = (8,8))

my_cmap = matplotlib.cm.get_cmap('Reds')
my_cmap.set_under('yellowgreen')

im = ax.imshow(np.log10(matrix_heatmap), cmap = my_cmap, vmin = 1e-5, vmax = 3)

ax.set_xticks(calculate_ticks(param_ticks['a1'], [7.7,10], [0,799]))
ax.set_xticklabels(param_ticks['a1'], fontsize = 14)
ax.set_xlabel('Clearance rate $a_1$', fontsize = 15)

ax.set_yticks(799-calculate_ticks(param_ticks['m'], [1500,25000], [0,799]))
ax.set_yticklabels(['1,500','5,000','10,000','15,000','20,000','25,000'], fontsize = 14)
ax.set_ylabel('Number of lethal positions $m$', fontsize = 15)

cbar = plt.colorbar(im, fraction=0.046, pad=0.04, ticks = [1e-5,1,2,3])
cbar.ax.tick_params(labelsize=15)
cbar.ax.set_yticklabels(['1','10','100','>1000']) 

#plt.savefig('treatment_peak_fullmrange_FEB1bis.jpg', dpi = 400, bbox_inches ='tight')
```

## Reduced range for m

```python
files_peak_short = [file for file in os.listdir('treatment_peak_short') if 'results' in file]
```

```python
a1values = list(np.linspace(7.7,10,800))
mvalues = list(np.linspace(10000,25000,800))[::-1]

param_ticks = {'m':[10000,12500,15000,17500,20000,22500,25000],
               'a1':[7.7,8.0,8.3,8.6,8.9,9.2,9.5,9.8,10]}
```

```python
matrix_heatmap_peak_short = np.zeros((len(mvalues), len(a1values)))

for find in range(len(files)):
    if find % 100 == 0:
        print(find)
    with open('treatment_peak_short/'+files[find], 'rb') as f:
        file_result = pickle.load(f)
        
    for r in file_result:
        mind = mvalues.index(r['m'])
        a1ind = a1values.index(r['a1'])
        matrix_heatmap_peak_short[mind, a1ind] =  r['u_ratio']
```

```python
def draw_altitude_line(matrix_heatmap, thres):

    altitude_line = []

    for a1ind in range(len(a1values)):
        array_search = matrix_heatmap[a1ind]
        index_alt = np.searchsorted(array_search, thres)
        if index_alt < len(a1values):
            altitude_line.append(index_alt)
        else:
            altitude_line.append(nan)
            
    return(altitude_line)
```

```python
fig, ax = plt.subplots(figsize = (8,8))

my_cmap = matplotlib.cm.get_cmap('Reds')
my_cmap.set_under('yellowgreen')

im = ax.imshow(matrix_heatmap_peak_short, cmap = my_cmap, vmin = 1.00001, vmax = 20)

ax.set_xticks(calculate_ticks(param_ticks['a1'], [7.7,10], [0,799]))
ax.set_xticklabels(param_ticks['a1'], fontsize = 14)
ax.set_xlabel('Clearance rate $a_1$', fontsize = 15)

ax.set_yticks(799-calculate_ticks(param_ticks['m'], [10000,25000], [0,799]))
ax.set_yticklabels(['10,000','12,500','15,000','17,500','20,000','22,500','25,000'], fontsize = 14)
ax.set_ylabel('Number of lethal positions $m$', fontsize = 15)

ax.plot(draw_altitude_line(matrix_heatmap_peak_short, 3), np.arange(800), c = 'b', linestyle = 'dashed')
ax.plot(draw_altitude_line(matrix_heatmap_peak_short, 10), np.arange(800), c = 'purple', linestyle = 'dashed')

cbar = plt.colorbar(im, fraction=0.046, pad=0.04, ticks = [1.00001,5,10,15,20])
cbar.ax.tick_params(labelsize=15)
cbar.ax.set_yticklabels(['1','5','10','15','>20']) 

#plt.savefig('treatment_peak_reducedmrange_FEB1bis.jpg', dpi = 400, bbox_inches ='tight')
```
