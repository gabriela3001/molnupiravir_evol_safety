```python
results_table_peak = []

a1_range = np.array([7.69,7.96,8.23,8.49,8.76])-2

for a1val in a1_range:
    sim_notreatment = simulation(init,b,a0,a1val,q0,q0,m,n,tim,5)
    sim_treatment = simulation(init,b,a0,a1val,q0,q1,m,n,tim,5)
    
    time_clearance = np.log(1e-4)/(b-a1val)
    
    YN = np.sum(sim_notreatment[:,1])*1e-3/1e3
    YN_round = np.round(YN, 2)
    print('YN', "{0:03}".format(YN))
    
    YM = np.sum(sim_treatment[:,1])*1e-3/1e3
    YM_round = np.round(YM, 2)
    print('YM', "{0:03}".format(YM))
    
    ERF = np.sum(sim_treatment[:,1])/np.sum(sim_notreatment[:,1])
    
    print('ERF', np.round(ERF,3))
    
    
    VN = np.sum(sim_notreatment)*1e-3/1e9
    VN_round = np.round(VN, 2)
    print('VN', "{0:03}".format(VN))
    
    VM = np.sum(sim_treatment)*1e-3/1e9
    VM_round = np.round(VM, 2)
    print('VM', "{0:03}".format(VM))
    
    ratio_V = np.sum(sim_treatment)/np.sum(sim_notreatment)
    ratio_V_rounded = np.round(ratio_V, 3)
    print('Vm/Vn', np.round(ratio_V, 2))
    print('\n')
    
    results_table_peak.append({'a1': a1val, 'T_clear':int(time_clearance),
                               'YN': YN_round, 'YM':YM_round,
                               'ERF':np.round(ERF,3), 
                               'VN':VN_round, 'VM':VM_round, 
                               'ratio_V':ratio_V_rounded})
```
